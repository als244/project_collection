```python
def my_tanh_activ(in_arr):
    rows = in_arr.shape[0]
    cols = in_arr.shape[1]
    activ = np.empty((rows, cols))
    for i in range(rows):
        for j in range(cols):
            val = in_arr[i, j]
            activ[i, j] = 1.7159 * math.tanh((2.0 / 3) * val)
    return activ
```

```python
def my_softmax(in_arr):
    classes = in_arr.shape[0]
    batch_size = in_arr.shape[1]
    out = np.empty((classes, batch_size))
    for samp in range(batch_size):
        s = 0
        for c in range(classes):
            s += np.exp(in_arr[c, samp])
        for c in range(classes):
            out[c, samp] = np.exp(in_arr[c, samp]) / s
    return out
```

```python
def my_softmax_jacobian(in_arr):
    classes = in_arr.shape[0]
    batch_size = in_arr.shape[1]
    jacobian = np.empty((batch_size, classes, classes))
    for samp in range(batch_size):
        for i in range(classes):
            for j in range(classes):
                if i == j:
                    jacobian[samp, i, j] = in_arr[i, samp] * (1 - in_arr[i, samp])
                else:
                    jacobian[samp, i, j] = -1 * in_arr[i, samp] * in_arr[j, samp]
    return jacobian
```

```python
def my_activation_deriv_soft(gradient, jacobian):
    classes = gradient.shape[0]
    batch_size = gradient.shape[1]
    activ_deriv = np.empty((classes, batch_size))
    for samp in range(batch_size):
        for i in range(classes):
            s = 0
            for k in range(classes):
                s += gradient[k, samp] * jacobian[samp, i, k]
            activ_deriv[i, samp] = s
    return activ_deriv
```

```python
def softmax_second_deriv(in_arr):
    rows = in_arr.shape[0]
    cols = in_arr.shape[1]
    secDeriv = np.empty((rows, cols))
    for i in range(rows):
        for j in range(cols):
            secDeriv[i, j] = in_arr[i, j] * (1 - in_arr[i, j])
    return secDeriv
```
